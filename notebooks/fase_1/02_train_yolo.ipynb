# 02 — Entrenar YOLO11 (Fase 1)

Entrena el detector **YOLO11** (alumno) sobre el dataset auto-etiquetado por SAM3 (maestro) en `01_sam3_to_yolo_autolabel`.

- Modelo: `yolo11s.pt` (más nuevo de Ultralytics, balance velocidad/precisión).
- Dataset: `yolo_dataset/data.yaml` (split por video, anti-leak).
- Salida: `runs/yolo11s_futbot/` (pesos en `weights/best.pt`).

Para correr automático/desatendido: `python train_yolo.py`.

```python
# pip install ultralytics   # si falta
from pathlib import Path
from ultralytics import YOLO

FASE1 = Path.cwd()
DATA = FASE1 / 'yolo_dataset' / 'data.yaml'
print('data:', DATA, DATA.exists())
```

```python
model = YOLO('yolo11s.pt')
model.train(data=str(DATA), epochs=100, imgsz=960, batch=16, patience=25,
            device=0, project=str(FASE1/'runs'), name='yolo11s_futbot', exist_ok=True)
```

```python
m = model.val()
print('mAP50-95:', float(m.box.map))
print('mAP50   :', float(m.box.map50))
# pesos: runs/yolo11s_futbot/weights/best.pt
```
